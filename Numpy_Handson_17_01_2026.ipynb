**Create numpy arrays which have zero,one,two,three dimesnion.Also print the shape and dimension of the numpy array.**

```python
#0d array
arr=np.array(30)
print(arr)
print('Shape:',arr.shape)
print('Dimension',arr.ndim)
```

```python
#1d array
arr=np.array([1,2,3,4,5])
print(arr)
print('Shape',arr.shape)
print('Diemshion:',arr.ndim)
```

```python
#2d array
arr=np.array([[1,2,3],[4,5,6],[8,9,10],[111,112,113]])
print(arr)
print('Shape',arr.shape)
print('Dimension',arr.ndim)
```

```python
#3d array
arr=np.array([[[1,2,3],[4,5,6]],[[9,10,10],[20,11,19]],[[17,18,19],[20,21,222]]])
print(arr)
print('Shape:',arr.shape)
print('Dimension:',arr.ndim)
```
