**Loc Function:**

.loc[] stands for location based on labels.

It is used to access rows or columns in a dataframe using their names or index labels

```python
#df.loc[row_label,column_label]
```

```python
data={
    'Name':['Umesh','Lakshmi','Mallick','Amol','Siddarth'],
    'Age':[20,22,24,26,28],
    'Marks':[80,85,90,78,50]
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
#select one row using the row_label
df.loc[2]
```

```python
#access multiple rows using the row label
df.loc[[1,3]]
```

```python
#accesss specific row and specific column
df.loc[[0,3],['Name','Age']]
```

```python
#slicing by label(end is included)
df.loc[1:3]
```

```python
df.loc[1:3,'Name':'Age']  #slicing across row label and column label
```

```python
#conditional selection
#i want to select thhose rows where my mrks is greater than 90
df.loc[(df['Marks']>=90)&(df['Age']>20),['Name','Marks']]
```
