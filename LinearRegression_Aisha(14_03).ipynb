```python
#Dropping null values
# df.dropna(inplace=True) inplace: permanent changes
```

```python
df.columns
```

```python
for col in df.columns:
  if df[col].dtype == 'object':
    df[col]=df[col].fillna(df[col].mode()[0])
  else:
    df[col]=df[col].fillna(df[col].mean())
```

```python
df.isnull().sum().sum()
```

```python
#Duplicates

df.duplicated().sum()
```

```python
#If we had duplicates,
#df.drop_duplicates(inplace=True)
```

```python
#Outlier Visualization

for col in df.columns:
  if(df[col].dtype != 'object'):
    sns.boxplot(df[col])
    plt.xlabel(col)
    plt.show()

#if no boxplot is visible, avoid touching that column because dropping it
#might drop the whole column
```

```python
#OUTLIER REMOVAL

outlist=['bmi','past_consultations','Hospital_expenditure','Anual_Salary']

for col in outlist:
  Q1=df[col].quantile(0.25)
  Q3=df[col].quantile(0.75)

  IQR=Q3-Q1

  LB=Q1-1.5*(IQR)
  UB=Q3+1.5*(IQR)

  df=df[(df[col]>=LB) & (df[col]<=UB)]
```

```python
df.info()
```

```python
df
```

```python
#Label Encoding : convert object data into numerical data
#arranges data in alphabetical order and then gives numerical values
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LE=LabelEncoder()
```

```python
for col in df.columns:
  if (df[col].dtype == 'object'):
    df[col]=LE.fit_transform(df[col])

#fit: understands the pattern(here, which unique category is given what number)
#transform: transform the data according the pattern understood
```

#MODEL BUILDING

```python
#1. splitting the data into x(features/independent) and y(target/dependent)
#2. splitting the data into TRAINING and TESTING SET
#3. Model Initialization
#4. Train the model
#5. Prediction
#6. Evaluate
```

```python
x=df.iloc[:,:-1] #all rows and all columns except last
#x=df.drop('Charges')
x
```

```python
y=df['charges']
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.80,random_state=0)
#random_state: fixes a set of sequence of randomly selected rows
#splits the dataset into training(80%) and testing data(20%)
#x-train & test : Features(input data) for training and testing
#y-train & test : Target(output data) for training and testing
```

```python
x_train #Questions for training
```

```python
y_train #Answers for training
```

```python
x_test #Questions for testing
```

```python
y_test #Answer for testing
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()

model.fit(x_train,y_train) #understand the pattern for training input and output
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
result=pd.DataFrame(columns=['Actual','Predicted'])
result['Actual']=y_test
result['Predicted']=y_pred
result
```

```python
from sklearn.metrics import * #import * = import everything

#r2 score: represents the proportion of variance in the dependent variable(y) explained by
#independent variable(x)
#if score is closer to 1, its a good model
res=r2_score(y_test,y_pred)
res
```

```python
#Best fit line: is a straight line that models the trend in data by minimizing the sum of squared residuals(diff b/w observed and predicted values)
#"Goodness of Fit": the accuracy of the line is evaluated using the r2 value and residual analysis
```

```python
sns.regplot(x=y_pred, y=y_test)
plt.title("Regression Plot")
plt.xlabel("Model Predictions")
plt.ylabel("Actual answers")
plt.show()
```
