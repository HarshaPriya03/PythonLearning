Ques 1- You are designing a simple banking management system. There is a base class called Account  that stores the attributes such as acc_no and balance. In this base class create  one method which will show the details - acc_no and blaance. From this base class create two child classes:
First child class -> SavingsAccount  -> method which will add 5% of interest and second Child class -> CurrentAccount -> method which will deduct 100 rs as a maintenance fee. Create object and show the demonstration.

```python
# 8 : 30 p.m
class Account:
  def __init__(self,acc_no,balance):
    self.acc_no = acc_no
    self.balance = balance

  def show_details(self):
    print('account number is -',self.acc_no)
    print('balance is -',self.balance)

class SavingsAccount(Account):
  def add_interest(self):
    self.balance = self.balance*1.05

class CurrentAccount(Account):
  def deduct_fee(self):
    self.balance -=100

s = SavingsAccount(101,50000)
print('-----------------------------------------------------------------------------------------------')
s.show_details()
print('------------------------------------------------------------------------------------------------------')
s.add_interest()
print('BALANCE AFTER ADDING INTEREST')
s.show_details()
print('------------------------------------------------------------------------------------------------')
c = CurrentAccount(110,56000)
c.show_details()
print('--------------------------------------------------------------------------------------------')
print('BALANCE AFTER DEDUCTING FEE')
c.deduct_fee()
c.show_details()
```

```python
class Account:
    def __init__(self,acc_no,balance):
        self.acc_no = acc_no
        self.balance = balance
    def show_details(self):
        print("Your acc_no",self.acc_no)
        print("Balance", self.balance)
class SavingsAccount(Account):
    def interest(self):
        self.balance *= 5
class CurrentAccount(Account):
    def maintenance(self):
        self.balance -= 100

o = SavingsAccount(101,5000)
o.show_details()
o.interest()
o.show_details()
c = CurrentAccount(1,250)
c.maintenance()
c.show_details()
```

**area()** , **perimeter()**

```python
# 8 : 46 p.m
class Rectangle:
  def __init__(self,length,width):
    self.length = length
    self.width = width

  def area(self):
    print("area of rectangle is -",self.length * self.width)

  def perimeter(self):
    print('perimeter of rectangle is -',2*(self.length+self.width))

r = Rectangle(20,30)
r.area()
r.perimeter()
```

Ques 3- Design a ride booking system with an abstract class Ride having an abstract method calculate_fare(). Create two child class BikeRide and CarRide which are inheriting from Ride class

```python
# 9 :04 p.m
from abc import ABC, abstractmethod
class Ride(ABC):
  @abstractmethod
  def calculate_fare(self,distance):
    pass

class BikeRide(Ride):
  def calculate_fare(self, distance):
    print('bike ride fare is -',distance*20)

class CarRide(Ride):
  def calculate_fare(self,distance):
    print('car ride fare is -',distance*40)

b = BikeRide()
c = CarRide()
b.calculate_fare(8)
c.calculate_fare(10)
```

```python
from abc import ABC, abstractmethod
class RideBooking(ABC):
    @abstractmethod
    def calculate_fare(self,kms):
        pass
class BikeRide(RideBooking):
    def calculate_fare(self,kms):
        print("Bike",kms*10)
class CarRide(RideBooking):
    def calculate_fare(self,kms):
        print("Car", kms*10)

bike=BikeRide()
bike.calculate_fare(2)

car = CarRide()
car.calculate_fare(3)
```

Ques 4 - Design a class BankAccount that has following attributes - acc_no , acc_holder , balance. Create these methods inside your class:

- deposit(amount)

- withdraw(amount)

- display_balance()

```python
# 9 :21 p.m
class BankAccount:
  def __init__(self,acc_no , acc_holder , balance):
    self.acc_no = acc_no
    self.acc_holder = acc_holder
    self.balance = balance

  def deposit(self,amount):
    self.balance += amount

  def withdraw(self,amount):
    if amount <= self.balance:
      self.balance -= amount
    else:
      print('insufficient balance')

  def show(self):
    print('balance is -',self.balance)

b = BankAccount(1234, 'Nisha',50000)
b.deposit(5000)
b.show()
b.withdraw(10000)
b.show()
```

Ques 5 - Create a Student Result System -

Create one class Student with attributes such as name , rollno , marks . This class will have a method grade() which will decide the grade of the student based on their marks - ( Grade A , Grade B , Grade C , else - "debard"). Show the functioning of the systema

```python
# 9 :37 p.m
class Student:
  def __init__(self,name,rollno , marks):
    self.name = name
    self.rollno = rollno
    self.marks = marks

  def grade(self):
    if self.marks >=90:
      print('Grade A')
    elif self.marks >=75:
      print('Grade B')
    elif self.marks >=55:
      print('Grade C')
    else:
      print('debard')

s = Student('Rishi',10,55)
s.grade()
```

Ques 7 - Create an abstract class Order with process_order() method. Child classes are -> OnlineOrder and StoreOrder. Your task here is to implement ths entire program

```python
# 10 :32 p.m
from abc import ABC , abstractmethod
class Order(ABC):
  @abstractmethod
  def process_order(self):
    pass

class OnlineOrder(Order):
  def process_order(self):
    print('processing online order:)')

class StoreOrder(Order):
  def process_order(self):
    print('processing store order:)')

o = OnlineOrder()
s = StoreOrder()
o.process_order()
s.process_order()
```
