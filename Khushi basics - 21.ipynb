**CONDITIONAL STATEMENTS**

used to make a decision based on some condition or criteria

- if - single choice/condition

- elif - multiple choice/conditions

- else

```python
# syntax-
# if (condition):
#   decision/body of if         -- it will execute when condition is true

# ---------------------------------------------------------------------------------------------------------------
# if-else
# if(condition):
#  decision/body of if           -- it will execute when condition is true
# else:
#  decision/body of else
# ------------------------------------------------------------------------------
# if-elif-else
# if (condition):
#   body/decision                -- it will execute when condition is true
# elif (condition):
#   body/decision                -- it will execute when elif condition is true
# elif (condition):
#   body/decision
# else:
#   body/decision
```

```python
# example for if-
m = int(input("enter your marks- "))
if (m>35):
  print("pass!!!")
```

```python
# using if-else:
m = int(input("enter your marks - " ))
if m>=40:
  print('pass')
else:
  print('fail')
```

```python
# if-elif-else
m = int(input("enter your marks - "))
if (m>=85):
  print("Grade A")
elif ((m>=75) and (m<85)):
  print('Grade B')
elif ((m>=65) and (m<75)):
  print('Grade C')
elif ((m>=55) and (m<65)):
  print('Grade D')
else:
  print('Grade E')
```

**2- While loop-** used to repeat a block of code as long as the condition is true. In while loop we don't know that how many times loop is going to repeat.

```python
# syntax - while (consition):
```

```python
5!=11
```

```python
secret = 11
guess = int(input("guess the number- "))
while (guess!=secret):
  print('Sorry, Wrong Answer:(')
  print('Try again!')
  guess = int(input('enter number again- '))
print('Correct Answer:)')
print('Level Unlocked!!!')
```

```python
password = 'vikesh'
a1 = str(input("enter password: "))
while(a1!=password):
  print('incorrect password')
  a1 = str(input('try again- '))
print('phone unlocked!! ')
```

CONTROL STATEMENT - used to control the flow or execution of our program.

- break = terminate/exit from the program

- continue = skip that value and continue to next

- pass = to avoid the error

```python
for i in range(1,21):
  if(i==15):
    break
  print(i)
```

```python
for i in range(1,21):
  if(i==15):
    continue
  print(i)
```

```python
for i in range(1,11):
  pass
```

```python
if 5>10:
  pass
```

**FUNCTIONS**

- It is a reusable block of code that performs a specific task or function when it is called

- define a function - def function_name():

- inside the body of function - logic/task

- call the function

```python
# function without parameter
def add():
  a = int(input('maths marks- '))
  b = int(input('science marks- '))
  print("total-" ,a+b)

# function calling is always done outside the function
# how to call? -> functionname
add()
```

```python
add()
```

```python
# function with parameter-
def add(a,b):
  print("total marks- ",a+b)

add(45,48)
n1 = int(input('num1- '))
n2 = int(input('num2- '))
add(n1,n2)
```

```python
# with default parameter-
def add(a=0,b=0):
  print('total -',a+b)

add()
add(88,80)
```

```python
def add(a,b=9):
  print('total -',a+b)

add(1)
```
