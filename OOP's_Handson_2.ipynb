Q1 - (a)- Create a shape class, with two abstract methods

  - area

  - perimeter

  (b) - Create a rectangle class , with variables length and breadth which is a child of shape

  (c) - create a circle class with variable as r , again it is a child of shape class


```python
# 8 :23 p.m
from abc import ABC, abstractmethod
class Shape(ABC):
  @abstractmethod
  def area(self):
    pass

  @abstractmethod
  def perimeter(self):
    pass

class Rectangle(Shape):
  def __init__(self , length , breadth):
    self.length = length
    self.breadth = breadth

  def area(self):
    print("area is -",self.length * self.breadth)

  def perimeter(self):
    print("perimeter is: ",2 * (self.length + self.breadth))

class Circle(Shape):
  def __init__(self , r):
    self.r = r

  def area(self):
    print("area is - ",(22/7)*self.r**2)

  def perimeter(self):
    print("perimeter is -",2*(22/7)*self.r)


r = Rectangle(10,20)
r.area()
r.perimeter()
c = Circle(7)
c.area()
c.perimeter()
```

Q4- Library Management System:- create a class Library

 - Initially we have a empty list of books

 - Have methods to add book , see all the books , and issue the book

 - You can only issue a book if it is present in the library (the list of books)

```python
# 9 :23 p.m
class Library:
  def __init__(self):
    self.Books = []

  def add(self,book):
    self.Books.append(book)

  def see(self):
    if self.Books:
      print('Books in the library are -')
      print(self.Books)
    else:
      print("No books are available")

  def issue(self,book):
    if book in self.Books:
      print(book , ' is issued')
      self.Books.remove(book)
    else:
      print(book , ' is not available')
```

```python
b1 = Library()
b1.see()
b1.add('Physics')
b1.add('Chemistry')
b1.add('Harry Potter')
b1.add('Python')
b1.see()
b1.issue('Jungle Safari')
b1.issue('Python')
b1.see()
```

Q5 - Create a class HostelRoom:

- attributes -> room_number , capacity , occupied_bed(initially 0)

- methods -> allocate_bed (can't allocate if the capacity is full)

- vacate_bed (decrease the number of occupied)

- room_status() print the details of the room

```python
# 10 :08 p.m
class HostelRoom:
  def __init__(self , room_number , capacity):
    self.room_number = room_number
    self.capacity = capacity
    self.occupied = 0

  def allocate(self):
    if self.occupied < self.capacity:
      print("Allocated!")
      self.occupied +=1
    else:
      print("Full!")

  def vacate(self):
    if self.occupied>0:
      print("vacated!")
      self.occupied -=1
    else:
      print("no beds occupied")

  def status(self):
    print("capacity is -",self.capacity)
    print("occupied is - ",self.occupied)
```

```python
r1 = HostelRoom(1,5)
r1.status()
r1.allocate()
r1.status()
r1.vacate()
r1.status()
```

Q6 - Create a student class with (name , rollno , marks)

- details() print all the details

- create 5 students , print the top 3 students based on the marks

Hint -> Sort objects based on the marks and then print top 3

```python
# 10 : 31 p.m
class Student:
  def __init__(self , rollno , name , marks):
    self.rollno = rollno
    self.name = name
    self.marks = marks

  def details(self):
    print('name is -',self.name , ', and roll no is- ',self.rollno)
    print("marks-",self.marks)

s1 = Student(1 , 'Aditya',85)
s2 = Student(2, 'Rhea',67)
s3 = Student(3, 'Sarthak',89)
s4 = Student(4, 'Krishna',54)
s5 = Student(5, 'Saiyam',78)
```

```python
students = [s1,s2,s3,s4,s5]
stored_stu = sorted(students, key=lambda s: s.marks , reverse=True)
for i in stored_stu[:3]:
  print(i.name ,i.marks)
# key=lambda s: s.marks -> sort using marks
# reverse = true > highest to lowest (descending)
```
