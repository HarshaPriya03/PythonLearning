```python
df=sns.load_dataset('titanic')
```

```python
df
```

a) Is there any relationship between Age and Fare,and does it diifer by survival status.

```python
survived = df[df["survived"] == 1]
not_survived = df[df["survived"] == 0]

plt.scatter(survived["age"], survived["fare"], label="Survived")
plt.scatter(not_survived["age"], not_survived["fare"], label="Not Survived")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Age vs Fare by Survival Status")
plt.legend()
plt.show()
```

# Insight:

There is no strong linear relationship between age and fare, but higher fares show a higher proportion of survivors.

Age does not strongly determine fare, but survival is clearly associated with higher fares.


b)How does the survival count differe between males and females?

```python
sns.countplot(x="sex", hue="survived", data=df)
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Survival Count by Gender")
plt.show()
```

# **Insight:**

Females survived in higher numbers than males

female passengers had a much higher survival rate than male passengers.



c) What is the age distribution of passengers on the Titanic?

```python
sns.histplot(df["age"].dropna(), bins=30)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution of Passengers")
plt.show()
```

# Insight:

Most passengers were young to middle-aged adults.

Most passengers were young adults, with fewer children and elderly passengers.
