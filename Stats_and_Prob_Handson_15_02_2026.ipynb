**Q5.Plot all the columns using a boxplot and detect the outliers and then  Write a function to remove the outliers in the data?**

```python
def outlier_remover(df):
  for i in df.columns: #for each and every column in df.columns
   if df[i].dtype!='object': #if data type is not object-Numerical columns
     Q1=df[i].quantile(0.25)
     Q3=df[i].quantile(0.75)
     IQR=Q3-Q1
     UL=Q3+1.5*IQR
     LL=Q1-1.5*IQR
     df=df[(df[i]>=LL) & (df[i]<=UL)]
```

```python
outlier_remover(df)
```
