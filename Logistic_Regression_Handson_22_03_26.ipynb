```python
# Detect the Outliers

# time -> 8:35
```

```python
import matplotlib.pyplot as plt
```

```python
for i in df.columns:
  if(df[i].dtype!="object"):
    plt.boxplot(df[i])
    plt.title(i)
    plt.show()
```

```python
outliers = ["age", "duration", "campaign", "cons.conf.idx"]

# Remove outliers from above columns
```

```python
# time -> 8:51
```

```python
for i in outliers:
  Q1 = df[i].quantile(0.25)
  Q3 = df[i].quantile(0.75)
  IQR = Q3-Q1
  LB = Q1 - 1.5*IQR
  UB = Q3 + 1.5*IQR
  df = df[(df[i] >= LB) & (df[i] <= UB)]
```

```python
for i in df.columns:
  if(df[i].dtype != "object"):
    plt.boxplot(df[i])
    plt.title(i)
    plt.show()
```

```python
confusion_matrix(y_test, y_pred)
```

```python
# [[TN  FP]
# [FN  TP]]
```

```python
# TN(True Negative) -> Actual value is Neg and Predicted Neg
# FP(False Positive) -> Actual -> Neg and Predicted -> Pos
# FN(False Negative) -> Actual -> Pos and Predicted -> Neg
# TP(True Positive) -> Actual -> Pos and Predicted -> Pos
```

```python
# Accuracy -> (TP + TN) / (TP+FP+TN+FN)
```

```python
(202 + 6369) / (202 + 122 + 6369 + 341)
```

```python
print(classification_report(y_test, y_pred))
```

```python
x_test.shape
```

```python
confusion_matrix(y_test, y_pred)
```

```python
# Precision -> How many times your model is actually correct among all the time it is saying as correct
# 0 -> No -> Negative
# Actual -> Neg , Model -> Neg   -> TN -> 6369

# Actual -> Pos  , Model -> Neg  -> FN -> 341
```

```python
# Confidence
# TN / TN+FN
(6369) / (6369+341)
```

```python
# for class 1 -> Yes -> Positive
# Precision ->
# Actual -> Pos  , model -> Pos  => TP
# Actual -> Neg  , model -> Pos  => FP
# Precision = TP / (TP+FP)
202 / (202+122)
```

```python
# Recall -> How many times your model is Neg. among how many time the value is neg.
# Actual -> Neg , Model -> Neg  -> TN
# Actual -> Neg , Model -> Pos  -> FP

# TN / (TN+FP)
```

```python
6369 / (6369+122)
```

```python
# F1 Score -> Balance between your precision and recall -> Harmonic Mean
```

```python
# 2 * [(Precision*recall) / (Precision+Recall)]
```

```python
2*((94.9*98.1)/(94.9+98.1))
```
