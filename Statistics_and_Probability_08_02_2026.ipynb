# **It is divided into two categories:**

1. **Descriptive Statistics**

2. **Inferential Statistics**

**1. Descriptive Statistics:**

To get some information from the entire data is called as descriptive statistics.It summarise and describe the features of  dattaset.

EX: If a Teachers calculates the average test score of the class that is descriptive statistics.

**Descriptive Statistics is divided into two parts:**

**1. Measure of Central Tendency:**

a) Mean

b) Median

c) Mode

**2. Measure of Dispersion:**

a) Range

b) Variance

c) Standard Deviation

d) Percentile

e) Quartiles

f) IQR

g) Outliers

h) Correlation

i) Normal Distribution

j) Skewness

k) Kurtosis

**2. Inferential Statistics:**

# **Variance:**

```python
data=[58,62,60,61,59]
var_data=np.var(data)
var_data
```

If my data Points are Closer to each other we will get the less variance as well as standard deviation.
