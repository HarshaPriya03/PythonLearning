Ques 5- Write a function that returns the factorial of a number-

```python
# Multiplying the number by all whole numbers smaller that it , till 1
# ex - 7! = 7 * 6 * 5 * 4 * 3  * 2 * 1
```

```python
# 9 : 16 p.m
def factorial(n):
  a = 1
  for i in range(1,n+1):
    a = a*i
  print(a)

b = int(input('enter number -'))
factorial(b)
```

Ques 7- Check whether a number is an armstrong number

```python
# An armstrong number is a number where the sum of each digit raised to the power of total digits is equal to the original number
```

```python
# ex - 153
# total digit = 3
# digits = 1             5           3
#       (1)^3 + (5)^3 + (3)^3
#         1+ 125 + 27 => 153
# original number = 153
```

```python
# 10 : 25 p.m
num = int(input('enter the number-'))
temp = num
digits = len(str(num))
sum = 0

# each digit raise to the power of total digit
while temp>0:
  last = temp % 10  # extract the last digit of the number
  sum = sum + (last**digits)
  temp = temp//10       # remove the last digit


if num == sum:
  print("armstrong number")
else:
  print("not an armstrong number")
```

Ques 9- Create a tuple with different numbers, count how many even and odd numbers are present in it.

```python
# 10 : 58 p.m
t = (10,21,22,13,33,29,30,40)
ecount = 0
ocount = 0

for i in t:
  if i%2==0:
    ecount+=1
  else:
    ocount +=1

print("Even count-",ecount)
print("odd count-",ocount)
```
