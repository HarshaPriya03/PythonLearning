```python
mat1 = np.random.randint(10, 51, size = (3,3))
mat2 = np.random.randint(50, 100, size = (3,3))
```

```python
mat1
```

```python
mat2
```

```python
#addition of matrices
mat1 + mat2 #np.add(mat1, mat2)
```

```python
mat1 + 10
```

```python
#subtraction of matrices
mat1 - mat2
```

```python
mat2 - mat1
```

```python
#division of matrices
mat1 / 10
```

```python
mat1 / mat2
```

```python
#element-wise multiplication or Hadamard product
mat1 * mat2
```

```python
#actual matrix multiplication : Dot product
mat1 @ mat2 #np.dot(mat1, mat2)
```

```python
#transpose of a matrix
mat1.T
```

```python
#matrix rank : helps us to identify duplicate or redundant data
np.linalg.matrix_rank(mat1)
```

```python
#determinant of a matrix : whether a matrix is singular or not
np.linalg.det(mat1)
```

```python
matrix = np.array([[3, -2],
                   [1, 0]])
```

```python
eigenvalues, eigenvectors = np.linalg.eig(matrix)

print("Original Matrix: ")
print(matrix)
print('\nEigenvalues: ')
print(eigenvalues)
print('\nEigenvectors: ')
print(eigenvectors)
```

```python
np.dot(matrix, [0.89442719, 0.4472136])
```
