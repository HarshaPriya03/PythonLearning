**1- INHERITANCE-** It is a property of object oriented programming in which a class inherits(extracts/takes) the property of another class.

   - the class which inherits the property ------------**child class / derived class**

   - the class from which we inherit the property ------------**parent class / base class**

```python
# WHY USE INHERITANCE?
# Reusability  -> write common code once in parent
```

```python
# TYPES OF INHERITANCE -
# 1- Single Level Inheritance
# 2- Multilevel Inheritance
# 3- Multiple Inheritance
# 4- Hierarchical Inheritance
# 5- Hybrid Inheritance
```

**(a)- Single Inheritance-** It is a type of inheritance in which one child class will inherit from one parent class

```python
class Animal:
  def speak(self):
    print('animal can make sounds')

class Dog(Animal):
  def bark(self):
    print('dogs can bark')

d = Dog()
d.bark()
d.speak()
```

**(b) - Multilevel Inheritance-** It is a type of inheritance in which a class inherits from another class and another class inherits from the derived class.

image.png

```python
class Animal:
  def eat(self):
    print('eating!!!!!')

class Mammal(Animal):
  def walk(self):
    print('walking!!!!')

class Dog(Mammal):
  def bark(self):
    print('bark!!!')

d = Dog()
m = Mammal()
d.walk()
d.eat()
m.eat()
d.bark()
# m.bark() - error , because m is an object of parent class for dog
```

**(c)- Multiple Inheritance-** It is a type of Inheritance in which a class inherits from more than one parent class.

image.png

```python
class Father:
  def play(self):
    print('father loves to play tennis')
class Mother:
  def cook(self):
    print('mother loves to cook')
class Son(Father,Mother):
  def sing(self):
    print('child loves to sing')

s = Son()
s.play()
s.cook()
```

**(d)- Hierarchical Inheritance-** It is a type of Inheritance in which more than one child class will inherit from one parent class

```python
class Animal:
  def eat(self):
    print('eating!!')
class Dog(Animal):
  def bark(self):
    print('bark!!!')
class Cat(Animal):
  def meow(self):
    print('meow!!!')

d = Dog()
c = Cat()
d.eat()
c.eat()
```

**(e) - Hybrid Inheritance-** It is a type of Inheritance in which we are having combination/mixture of two or more types of inheritance

(ex- single inheritance + multiple inheritance)

```python
class A:
  def fun1(self):
    print('class A')

class B(A):
  def fun2(self):
    print('class B')

class C:
  def fun3(self):
    print('class C')

class D(B,C):
  def fun4(self):
    print('class D')

d = D()
d.fun3()
d.fun2()
d.fun1()
```

**2- ENCAPSULATION-** It means wrapping the data (variables) and methods (functions) together into one single unit (class) - and keeping some information private or secure.

- let's think of a capsule:

- A capsule hides the bitter medicine inside and we are only able to see the outer cover

- Similarly, in terms of programming, encapsulation hides the internal details of how data works and show what's necessary.

```python
# HOW ENCAPSULATION HIDES THE DATA?
```

Encapsulation is done in Python using Access modifiers-

- Access Modifiers -> These are used to control the access

- There are three types of Access modifiers 

- **1- Public-** accessible from anywhere (i.e- inside or outside the class)

- created normally------> ex - **name** -> **self.name**

- No restriction with public data members , ex - (your account name)

- **2- Protected-** accessible within the class and child class

- created using single underscore ---> ex - **salary** -> **self._salary**

- Medium restriction , ex - (your account balance-) (visible to bank)

- **3- Private-** accessible only inside the class

- created using double underscore ---> ex - **balance** -> **self.__balance**

- High restriction , ex - (Your ATM pin) (only you know)

```python
# public access modifier -
class Car:
  def __init__(self,color,brand):
    self.color = color # public
    self.brand = brand

c = Car('Black','BMW')
print(c.color)
print(c.brand)
```

```python
# protected access modifier -
class Employee:
  def __init__(self,name , salary):
    self.name = name # public
    self._salary = salary  # protected

class Manager(Employee):
  def show_salary(self):
    print('salary is -',self._salary)

m = Manager('Rishi', 45000)
m.show_salary()
```

```python
class Employee:
  def __init__(self,name , salary):
    self.name = name # public
    self._salary = salary  # protected

class Manager(Employee):
  def show_salary(self):
    print('salary is -',self._salary)

class Hi():
    def hello():
        print("salary",self._salary)

m = Manager('Rishi', 45000)
m.show_salary()
h=Hi()
h.hello()
```

```python
# private access modifier-
class BankAccount:
  def __init__(self,balance):
    self.__balance = balance

  def deposit(self,amount):
    if amount>0:
      self.__balance +=amount

  def show_balance(self):
    print('balance is -',self.__balance)

o1 = BankAccount(30000)
o1.deposit(20000)
o1.show_balance()
# print(o1.self__balance) -> can't be accessed outside the class
```

```python
class BankAccount:
  def __init__(self,balance):
    self.__balance = balance

  def deposit(self,amount):
    if amount>0:
      self.__balance +=amount

  def show_balance(self):
    print('balance is -',self.__balance)

class SBI(BankAccount):
  def show(self):
    print('original balance is -',self.__balance)

s = SBI(30000)
#s.show()    private data member - can't be accessed inside the child class
```

**3- POLYMORPHISM-** "many forms"

- Poly = many

- morphism = form

- The same thing behaves differently in differnt situation

```python
# In terms of programming -
# It allows us to use the same function/method name in different ways,
# depending on the object or class that is using it
```

```python
# TWO TYPES OF POLYMORPHISM-
# 1- Method Overloading
# 2- Method Overriding
```

**1- Method Overloading-** When one class has multiple methods with the same name but different numbers of parameters or different types of parameters.

```python
# In python method overloading is not directly supported
# if we try to do method overloading , it will override the result
```

```python
class Demo:
  def show(self,a):
    print('one argument:-',a)

  def show(self,a,b):
    print('two argument:-',a,b)

d = Demo()
# d.show(10) error - latest function with 2 arguments is considered
d.show(10,17)
```

```python
# but we can stimulate method overloading using default arguments
```

```python
class calculator:
  def add(self,a=0,b=0,c=0):
    print(a+b+c)

c = calculator()
c.add(10,11)
c.add(2,3,4)
c.add(8)
```

**2- Method Overriding-** When a child class has a method with the same name as a method present in the parent class, and it changes or override the parent's method behaviour

```python
class Animal:
  def sound(self):
    print('can make some sounds')

class Dog(Animal):
  def sound(self):
    print('bark')

d = Dog()
d.sound()
```

```python
# super() - It is a keyword used to call the parent class method inside the child class
```

```python
class Animal:
  def sound(self):
    print('can make some sounds')

class Dog(Animal):
  def sound(self):
    print('bark')
    super().sound()

d = Dog()
d.sound()
```

```python
# ques - write a program for displaying method overloading
# create a class Camera with a method take_photo(timer=0, flash=False) that can take a photo :
# instantly without flash,
# after a given timer without flash
# after a given timer with flash
```

```python
# 9 : 37 p.m
class Camera:
  def take_photo(self,timer=0,flash=False):
    if timer==0 and not flash:
      print('photo taken instantly!:)')
    elif timer >0 and not flash:
      print(f'photo will be taken after {timer} seconds:) ')
    elif timer >0 and flash:
      print(f'photo will be taken after {timer} seconds with flash ON:)')
    else:
      print('photo done:)')
c = Camera()
c.take_photo()
c.take_photo(timer=3)
c.take_photo(timer=5 , flash=True)
```

**What is an Abstract Class?**

- A class that is incomplete and cannot be used directly (cannot create an object of an abstract class)

- An abstract class is a blueprint for other classes

- it can have normal methods and abstract methods

**What is an Abstract Method?**

- It is a method which is created inside abstract class

- A method that has name only , no code/no body

```python
# from where we get abstract class and abstract method?
```

```python
from abc import ABC , abstractmethod
# abc -> module for abstract base class
# ABC -> Abstract class
```

```python
# how to create an abstract class
class Person(ABC): # here person has become abstract class
  @abstractmethod
  def role(self):
    pass
class Student(Person):
  def role(self):
    print('role is to study')

class Teacher(Person):
  def role(self):
    print('role is to teach')

s= Student()
t = Teacher()
s.role()
t.role()
```

```python
from abc import ABC , abstractmethod
class Payment(ABC):
  @abstractmethod
  def pay(self,amount):
    pass

class Cash(Payment):
  def pay(self,amount):
    print(f'paid {amount} using cash')

class Card(Payment):
  def pay(self,amount):
    print(f'paid {amount} using card')

c1 = Card()
c2 = Cash()
c1.pay(500)
c2.pay(1000)
```

- class says here that any payment must have a pay() method -> (only rule , no logic , no print)
