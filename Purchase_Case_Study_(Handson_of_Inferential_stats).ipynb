# Question 1


# It was observed that the **average purchase of Gender = Men and Age group = 18-25 was 10,000** is it correct or not ?

```python
# Without any inferential stats lets check whether average purchase is 10k or not

# Fetch the data where Gender==Male , Age ==18-25
new_data = df[ (df["Gender"]==1) & (df["Age"]=="18-25") ]
# Now find the average purchase amount directly
print("Average purchase amount is :",new_data["Purchase"].mean() )
```

```python
# Now lets do the same thing by using Inferential statitis test's
```

**A one-sample t-test** is a statistical test used to determine whether the mean of a single sample is significantly different from a known or hypothesized population mean.

It helps in testing hypotheses about the average value of a dataset.

**Types of t-tests:**

**One-sample t-test**: Compares the sample mean to a known value (population mean).

**Independent two-sample t-test**: Compares the means of two independent groups.

**Paired sample t-test**: Compares means from the same group at different times (e.g., before and after treatment).

Key Concepts:

Null Hypothesis (H₀): The sample mean is equal to the population mean.

Alternative Hypothesis (H₁): The sample mean is different (greater or less) than the population mean.

```python
# Prepare a sample of 10% of full data  ,Fetch the data where Gender==Male , Age ==18-25
new_data = df[ (df["Gender"]==1) & (df["Age"]=="18-25") ].sample(3600)

# Lets Create 2 hypothesis
'''
Null hypothesis        = Average purchase is 10,000
Alternate hypothesis  = Average purchase is not 10,000
'''

# Apply the appropriat test that is 1 sample t test
from scipy.stats import ttest_1samp
t_stats , p_value = ttest_1samp( new_data["Purchase"]  , 10000)
print(p_value)
# Compare the p value
'''
if p value is less than 0.05 = Accept the Alternate hypothesis
if p value is more than 0.05 = Accept the Null hypothesis
'''
print("Exact p value is :", f"{p_value:.25f}")  # .25f means upto 25 decimal place fix the value
```

**Our p value is less than 0.05 so that means we accept the Alternate hypothesis ( Average purchase is not 10,000 )**

# It was observed that the percentage of women who spend more than 10,000 is 35% out of all people.

You will use a **one-sample z-tes**t for proportions to compare the observed proportion of women spending more than 10,000 with the hypothesized proportion of 35%.

**Why Use a One-Sample Z-Test for Proportions?**

A z-test for proportions is used when you want to compare an observed sample proportion to a known or hypothesized population proportion.

```python
# Lets check by using descriptive stats

# Lets check total people who spend more than 10k out of them how much poercent are women
new_data = df[df["Purchase"]>10000]
print(new_data.shape)
# So 91057 people are spending more than 10,000

# Lets find how many of them are women
new_data = df[ (df["Purchase"]>10000) & (df["Gender"]==0) ]
print(new_data.shape)
# so 18685 are women who spend more than 10k

# Lets find the percenatge of women
print("How much % of women spend more than 10k :" , (18685/91057)*100)
```

```python
# Lets check by using Inferential  stats

# Lets check total people who spend more than 10k out of them how much poercent are women
#total_people = df[df["Purchase"]>10000]
total_people=91057

# Lets find how many of them are women
#women_count = df[ (df["Purchase"]>10000) & (df["Gender"]==0) ]
women_count=18685
# Hypothesis value
p0 = 0.35

from statsmodels.stats.proportion import proportions_ztest
z_stats , p_value = proportions_ztest ( count =women_count  , nobs=total_people , value =p0  )
print(p_value)
# Compare the p value
'''
if p value is less than 0.05 = Accept the Alternate hypothesis ( That count of women who spend more than 10k is not 35%)
if p value is more than 0.05 = Accept the Null hypothesis ( That count of women who spend more than 10k is 35%)
'''
print("Exact p value is :", f"{p_value:.25f}")  # .25f means upto 25 decimal place fix the value
```

```python
# Inferential Stats

from statsmodels.stats.proportion import proportions_ztest

df_male = df[df['Gender'] == 1]   # 1 = Male
df_male['High_Spend'] = df_male['Purchase'] > 10000

age_18_25 = df_male[df_male['Age'] == '18-25']
age_26_35 = df_male[df_male['Age'] == '26-35']


count = [age_18_25['High_Spend'].sum(),age_26_35['High_Spend'].sum()]
nobs = [len(age_18_25),len(age_26_35)]

z_stat, p_value = proportions_ztest(count, nobs)
'''
if p value is less than 0.05 = Accept the Alternate hypothesis ( Percentages are SAME.)
if p value is more than 0.05 = Accept the Null hypothesis ( Percentages are DIFFERENT)
'''
print("P-Value:", p_value)
```
