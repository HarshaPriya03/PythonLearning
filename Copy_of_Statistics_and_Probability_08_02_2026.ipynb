# **IQR(Interquartile Range)**

image.png

# Steps to Find the outliers:

Calculate the Q1 and Q3.

Find IQR(Interquartile Range).

Find the Upper and Lower Limit.

If any value is higher than Upper Limit and Lower than Lower Limit will be the outlier.

```python
age=[18,25,30,45,33,67,78,93,250,76,99,271,-90]
age
```

```python
Q1=np.percentile(age,25)
Q1
```

```python
Q3=np.percentile(age,75)
Q3
```

```python
IQR=Q3-Q1
```

```python
IQR
```

```python
UL=Q3+1.5*IQR
LL=Q1-1.5*IQR
```

```python
UL
```

```python
LL
```

```python
#187.5- If any value will be higher than 187.5 will be an outlier
#-64.5- If any value will be lower than -64.5 will be an outlier
```

```python
for i in age:
  if i>UL or i<LL:
    print(f'Outliers value in thhe data is {i} and it has been detected')
```

```python
clean_age=[]
for i in age:
  if i>UL or i<LL:
    print(f'Outliers value in thhe data is {i} and it has been detected')
  else:
    clean_age.append(i)
```

```python
clean_age
```

```python
plt.boxplot(clean_age)
plt.show()
```

# Skewness:

How Data is Distributed:

skewnees=0-> Symmetric(Normal Distribution)

skewness>0-> Right Skewed(Long Right tail)

skewness<0-> Left Skewed(Long Left Tail)

```python
data=[10,11,11,12,12,12,12,12]
plt.hist(data,bins=3,color='pink',edgecolor='black')
plt.title('Left Skewed Data')
plt.show()
```

```python
skew(data)
```

```python
data=[10,10,10,10,10,15,20]
plt.hist(data,bins=3,color='pink',edgecolor='black')
plt.title('Right Skewed Data')
plt.show()
```

```python
skew(data)
```

2) Leptokurtic(High Kurtosis)

Kurtosis>3

Very sharp peak

Fewer extreme values
